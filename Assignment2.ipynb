```python
# 1. Write a user-defined function that takes a number as input and returns: 
# • "Positive Even" 
# • "Positive Odd" 
# • "Negative" 
# • "Zero" 
# Use proper conditional statements and explain your logic. 
# in python
# User-defined function to classify a number
def classify_number(num):
    
    # Check if the number is zero
    if num == 0:
        return "Zero"
    
    # Check if the number is negative
    elif num < 0:
        return "Negative number"
    
    # If not zero or negative, it must be positive
    else:
        # Check if the positive number is even
        if num % 2 == 0:
            return "Positive Even"
        # Otherwise, it is positive odd
        else:
            return "Positive Odd"


# Take input from the user with error handling
while True:
    try:
        number = int(input("Enter a number: "))
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")

# Call the function and store the result
result = classify_number(number)

# Print the result
print(result)
```

```python
# 2. Write a function that accepts three numbers and returns the largest number without using built-in 
# functions like max(). 
# Explain how your conditional structure ensures correctness. 

# User-defined function to find the largest of three numbers
def largest_number(a, b, c):

    # Assume first number is the largest initially
    largest = a

    # Compare second number with current largest
    if b > largest:
        largest = b

    # Compare third number with current largest
    if c > largest:
        largest = c

    # Return the final largest value
    return largest


# Take three numbers from the user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

# Call the function
result = largest_number(num1, num2, num3)

# Print the result
print("The largest number is:", result)
```

```python
# 3. Write a function that accepts a student’s marks and returns the grade based on the following logic: 
# • 90+ → A 
# • 75–89 → B 
# • 50–74 → C 
# • Below 50 → Fail 
# Use nested conditional statements and justify your approach. 
# User-defined function to calculate grade
def calculate_grade(marks):

    # First check if marks are 90 or above
    if marks >= 90:
        return "A"
    
    # If not A, check for B
    else:
        if marks >= 75:
            return "B"
        
        # If not B, check for C
        else:
            if marks >= 50:
                return "C"
            
            # If none of the above, student fails
            else:
                return "Fail"


# Take marks input from the user
marks = float(input("Enter student's marks: "))

# Call the function
grade = calculate_grade(marks)

# Display the grade
print("Grade:", grade)
```

```python
# 4. Write a function that takes age as input. 
# • If age is negative, raise an error. 
# • If age is below 18, return "Minor", otherwise return "Adult". 
# Explain how conditional statements help in input validation. 

# User-defined function to classify age
def check_age(age):

    # Validate input: age cannot be negative
    if age < 0:
        raise ValueError("Age cannot be negative.")
    
    # Check if age is below 18
    if age < 18:
        return "Minor"
    
    # If age is 18 or above
    else:
        return "Adult"


# Take input from user
try:
    age = int(input("Enter age: "))
    result = check_age(age)
    print(result)

# Handle possible errors
except ValueError as e:
    print("Error:", e)
```

```python
# 5. Write a function that accepts a list of integers and: 
# • Counts even numbers 
# • Ignores negative numbers 
# • Stops processing if zero appears 
# Use both loops and conditional statements inside the function. 
# # Function to count even numbers
def count_even_numbers(numbers):
    
    count = 0  # Variable to store count of even numbers
    
    # Loop through each number in the list
    for num in numbers:
        
        # Stop processing if zero appears
        if num == 0:
            print("Zero encountered. Stopping process.")
            break
        
        # Ignore negative numbers
        if num < 0:
            continue
        
        # Check if the number is even
        if num % 2 == 0:
            count += 1
    
    return count


# Example usage
numbers = list(map(int, input("Enter integers separated by space: ").split()))

result = count_even_numbers(numbers)

print("Count of even numbers:", result)
```
